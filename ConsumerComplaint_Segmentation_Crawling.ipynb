#Function to Scrape the Data

```
def Scrape(soup):
  Complaint_list =soup.find_all('table')[2:-2] # fetching all the complaints  from 3rd table to second last table as the last table and first two tables are of no use
  for complaint in Complaint_list:
    anchor_list =complaint.find_all('a')
    Complaint_Id.append(anchor_list[-5].get('id') )
    Complaint_Category.append(anchor_list[-1].get_text())
    Complaint_Text.append(complaint.find('div' , attrs={'style':'padding-top:6px;'}).get_text())
    desc =complaint.find_all('a')[-5].get_text()
    Company_Name.append(desc.split('—')[0])
    Complaint_Desc.append(desc.split('—')[-1])
```

```
Company_Name =[]
Complaint_Desc = []
Complaint_Text =[]
Complaint_Category =[]
Complaint_Id =[]
```

#Web Scrapping to fetch data from all the pages

```
with requests.Session() as session:
    page_number = 1
    url = 'https://www.consumercomplaints.in/?page='
    while True:
        print("Processing page: #{page_number}; url: {url}".format(page_number=page_number, url=url))
        response = session.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        Scrape(soup)

        # check if there is next page, break if not
        next_link = soup.find("a", text="Next")
        if next_link is None:
            break

        url = urljoin(url, next_link["href"])
        page_number += 1

print("Done.")
```
